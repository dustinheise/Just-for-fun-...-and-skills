```python
def bingo():
  print("Enter 'stop' to restart or end the game.\n")
  draw = []
  for i in range(1,75):
    r = rand.randint(1, 75)
    if r not in draw:
      draw.append(r)
    else:
      for i in range(1,75):
        r = rand.randint(1, 75)
        if r not in draw:
          draw.append(r)       
  #print(draw)
  for i in draw:
    if i >=1 and i<=15:
      print('B-' + str(i))
    
    elif i >= 16 and i <=30:
      print('I-' + str(i))

    elif i >= 31 and i <= 45:
      print('N-' + str(i))

    elif i >= 46 and i <=60:
      print('G-'+ str(i))

    elif i >= 61 and i <=75:
      print('O-' + str(i))
    
    next = input('press enter for next ')
    if next == 'stop':
      break
      

  restart = input("Enter 'y' to RESTART or 'n' to END ")
  if restart == 'y': 
    bingo()         
  else:
    print('\nThanks for playing') 
```
